```python
import pandas as pd
```

```python
trips = pd.read_csv("data/vista_trips_2023-2025.csv")
trips = trips[["year","origlga","destplace1","destplace2","destpurp1","destpurp2","destlga","triptime","cumdist"]]
trips = trips[(trips["year"] >= 2022) & (trips["destplace2"] == "Restaurant or Cafe") & (trips["destpurp2"] == "Ate or drank") & (trips["origlga"].str.endswith(" (C)"))]
trips = trips[~trips["origlga"].isin(["Warrnambool (C)", "Ballarat (C)", "Greater Geelong (C)", "Greater Bendigo (C)", "Greater Shepparton (C)"])]

trips["cumdist"] = pd.to_numeric(trips["cumdist"], errors="coerce")
trips["triptime"] = pd.to_numeric(trips["triptime"], errors="coerce")

trips["origlga"] = trips["origlga"].str.replace(" (C)", "", regex=False).str.upper()
trips
```

```python
cafe_trips = trips[["origlga","triptime","cumdist"]].groupby("origlga").mean()
cafe_trips = cafe_trips.sort_values("cumdist").reset_index()
cafe_trips.to_csv("data/cafe_trips_by_lga.csv", index=False)
cafe_trips
```
