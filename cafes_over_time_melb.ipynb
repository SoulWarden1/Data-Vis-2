```python
import pandas as pd
```

```python
cafe_data = pd.read_csv("data/cafes-and-restaurants-with-seating-capacity.csv")
cafe_data
```

```python
cafe_data = cafe_data[cafe_data["industry_anzsic4_description"] == "Cafes and Restaurants"]
cafe_data = cafe_data[["census_year","trading_name","business_address"]].drop_duplicates()
cafe_data
```

```python
cafes_by_year = cafe_data.groupby("census_year")['trading_name'].count()
cafes_by_year = cafes_by_year.rename("Number of Cafes")
cafes_by_year
```

```python
cafes_by_year.to_csv("data/cafes_by_year.csv")
```
