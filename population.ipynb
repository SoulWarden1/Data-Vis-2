```python
import pandas as pd
```

```python
data = pd.read_csv("data/2021Census_G01_VIC_SAL.csv")
data = data[["SAL_CODE_2021", "Tot_P_P"]]
data["SAL_CODE_2021"] = data["SAL_CODE_2021"].str.slice(3)
data
```

```python
suburb_names = pd.read_excel("data/SAL_2021_AUST.xlsx")
suburb_names
```

```python
suburb_names = suburb_names[suburb_names["STATE_NAME_2021"] == "Victoria"]
suburb_names = suburb_names[["SAL_CODE_2021", "SAL_NAME_2021"]].drop_duplicates()
suburb_names["SAL_NAME_2021"] = suburb_names["SAL_NAME_2021"].str.removesuffix(" (Vic.)").str.upper()
suburb_names
```

```python
data = data.merge(suburb_names, on="SAL_CODE_2021")
data = data[["SAL_NAME_2021", "Tot_P_P"]]
data = data.rename(columns={"SAL_NAME_2021": "Suburb", "Tot_P_P": "Population"})
data
```

```python
data.to_csv("data/population_by_suburb.csv", index=False)
```
